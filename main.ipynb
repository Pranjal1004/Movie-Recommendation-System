```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def extract_id(x, id):
    m_id = id
    result = tmdb_movie.search(x)
    if not result:
        # This "if not result" condition is specifically added for 2020 data as for some movie we don't get result on TMBb
        return np.NaN
    else:
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        # extracting data in json format
        data_json = response.json()
        # Actually there's a key "id" which basically holds the movie id and another key "imdb_id" 
        # which holds the movie's imdb id. 

        if m_id == "movie":
            # returns movie's id
            return data_json["id"]
        if m_id == "imdb":
            # returns movie's imdb id
            return data_json["imdb_id"]
        
        
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def create_similarity():
    data = pd.read_csv("movie_data_upto_2022_march.csv")
    cv = CountVectorizer()
    c_matrix = cv.fit_transform(data["combine"]+data["combine"])
    similarity = cosine_similarity(c_matrix)
    return data, similarity
```

```python
data, similarity = create_similarity()
```

```python
def recommend(m, data, similarity):
    m = m.lower()
    if m not in data.movie_title.unique():
        return print("Sorry, we dont have your movie.")
    else:
        data = data
        similarity = similarity
        i = data.movie_title.loc[data.movie_title == m].index[0]
        lst = list(enumerate(similarity[i]))
        lst = sorted(lst, key = lambda x : x[1], reverse=True)
        lst = lst[1:11]
        movies = []
        for i in range(len(lst)):
            movies.append(data.movie_title[lst[i][0]])
        return movies
```

```python
recommend("prisoners", data, similarity)
# see there are some movies which have same name due to which you may get the same movie in the recommended movies
```
